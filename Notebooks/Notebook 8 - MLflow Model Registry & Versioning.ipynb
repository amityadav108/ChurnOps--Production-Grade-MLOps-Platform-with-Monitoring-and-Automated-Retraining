### Imports

```python
import os
import shutil
import joblib
from datetime import datetime

import warnings
warnings.filterwarnings("ignore")
```

### Create Registry Structure

```python
os.makedirs("model_registry", exist_ok=True)

os.makedirs(
    "model_registry/staging",
    exist_ok=True
)

os.makedirs(
    "model_registry/production",
    exist_ok=True
)

print("Registry created successfully.")
```

### Register Version 1

```python
version = "v1"

version_path = os.path.join(
    "model_registry",
    version
)

os.makedirs(
    version_path,
    exist_ok=True
)

print(version_path)
```

### Save Metadata

```python
metadata = {
    "version": "v1",
    "created_at": str(datetime.now()),
    "model_name": "LightGBM",
    "status": "Registered"
}

print(metadata)
```

### Register Model

###  Promote to Staging

```python
shutil.copy(
    "models/best_model.pkl",
    "model_registry/staging/model.pkl"
)

print("Model moved to STAGING.")
```

### Promote to Production

```python
shutil.copy(
    "models/best_model.pkl",
    "model_registry/production/model.pkl"
)

print("Model moved to PRODUCTION.")
```

### Load Production Model

```python
production_model = joblib.load(
    "model_registry/production/model.pkl"
)

print(type(production_model))
```

```python
import joblib

X_test = joblib.load(
    "artifacts/preprocessing/X_test_processed.pkl"
)

print("X_test loaded successfully.")
```

### Test Production Model

```python
predictions = production_model.predict(
    X_test[:10]
)

print(predictions)
```

### Display Registry Structure

```python
for root, dirs, files in os.walk(
    "model_registry"
):

    level = root.replace(
        "model_registry",
        ""
    ).count(os.sep)

    indent = " " * 4 * level

    print(
        f"{indent}{os.path.basename(root)}"
    )

    sub_indent = " " * 4 * (level + 1)

    for file in files:
        print(
            f"{sub_indent}{file}"
        )
```

### Register Version 2 Example

```python
new_version = "v2"

os.makedirs(
    f"model_registry/{new_version}",
    exist_ok=True
)

shutil.copy(
    "models/best_model.pkl",
    f"model_registry/{new_version}/best_model.pkl"
)

print("Version 2 registered.")
```
