# Notebook 10 — Docker Containerization

This notebook containerizes the ChurnOps FastAPI application.

Goals:

- Create requirements.txt
- Create Dockerfile
- Create .dockerignore
- Build Docker image
- Run containerized API

## Linux Dependencies for LightGBM

LightGBM requires OpenMP libraries inside Docker containers.

Installed:

- gcc
- g++
- libgomp1

This ensures compatibility with Linux-based production environments.

### Imports


```python
import os
```

### Create Root requirements.txt

```python
requirements = """
fastapi
uvicorn
pandas
numpy
scikit-learn
lightgbm
joblib
pydantic
"""

with open(
    "requirements.txt",
    "w",
    encoding="utf-8"
) as f:
    f.write(requirements)

print("requirements.txt created successfully.")
```

### Create Dockerfile

```python
dockerfile = """FROM python:3.12-slim

WORKDIR /app

# Install system dependencies required by LightGBM
RUN apt-get update && apt-get install -y \
    gcc \
    g++ \
    libgomp1 \
    && rm -rf /var/lib/apt/lists/*

COPY requirements.txt .

RUN pip install --no-cache-dir -r requirements.txt

COPY . .

EXPOSE 8000

CMD ["uvicorn", "app.main:app", "--host", "0.0.0.0", "--port", "8000"]
"""

with open("Dockerfile", "w", encoding="utf-8") as f:
    f.write(dockerfile)

print("Dockerfile created successfully.")
```

### Create .dockerignore

```python
dockerignore = """
__pycache__/
.ipynb_checkpoints/
*.pyc
*.pyo
*.pyd
venv/
.env
mlruns/
notebooks/
"""

with open(
    ".dockerignore",
    "w",
    encoding="utf-8"
) as f:
    f.write(dockerignore)

print(".dockerignore created successfully.")
```

### Verify Files

```python
print(os.path.exists("Dockerfile"))
print(os.path.exists(".dockerignore"))
print(os.path.exists("requirements.txt"))
```

### Display Dockerfile

```python
with open("Dockerfile", "r") as f:
    print(f.read())
```
