### Imports

```python
import os
import mlflow
import joblib

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

### Close Previous Runs (IMPORTANT)

```python
print(mlflow.active_run())
```

### Create Experiment

```python
mlflow.set_experiment(
    "ChurnOps_MLOps_Experiment"
)

print("Experiment created successfully.")
```

###  Load Best Model

```python
model = joblib.load(
    "models/best_model.pkl"
)

print("Model loaded.")
```

### Load Processed Test Data

```python
X_test = joblib.load(
    "artifacts/preprocessing/X_test_processed.pkl"
)

y_test = joblib.load(
    "artifacts/preprocessing/y_test.pkl"
)

print("Test data loaded.")
```

### Generate Predictions

```python
predictions = model.predict(X_test)

probabilities = model.predict_proba(X_test)[:, 1]

print("Predictions completed.")
```

### Calculate Metrics

```python
accuracy = accuracy_score(
    y_test,
    predictions
)

precision = precision_score(
    y_test,
    predictions
)

recall = recall_score(
    y_test,
    predictions
)

f1 = f1_score(
    y_test,
    predictions
)

roc_auc = roc_auc_score(
    y_test,
    probabilities
)

print("Metrics calculated.")
```

### Display Metrics

```python
print("Accuracy :", round(accuracy, 4))
print("Precision:", round(precision, 4))
print("Recall   :", round(recall, 4))
print("F1 Score :", round(f1, 4))
print("ROC AUC  :", round(roc_auc, 4))
```

### Start MLflow Run


```python
with mlflow.start_run():

    # -------------------------
    # Parameters
    # -------------------------
    mlflow.log_param(
        "model_type",
        type(model).__name__
    )

    # -------------------------
    # Metrics
    # -------------------------
    mlflow.log_metric(
        "accuracy",
        accuracy
    )

    mlflow.log_metric(
        "precision",
        precision
    )

    mlflow.log_metric(
        "recall",
        recall
    )

    mlflow.log_metric(
        "f1_score",
        f1
    )

    mlflow.log_metric(
        "roc_auc",
        roc_auc
    )

    # -------------------------
    # Evaluation Artifacts
    # -------------------------
    if os.path.exists(
        "artifacts/evaluation/confusion_matrix.png"
    ):
        mlflow.log_artifact(
            "artifacts/evaluation/confusion_matrix.png"
        )

    if os.path.exists(
        "artifacts/evaluation/roc_curve.png"
    ):
        mlflow.log_artifact(
            "artifacts/evaluation/roc_curve.png"
        )

    if os.path.exists(
        "artifacts/evaluation/precision_recall_curve.png"
    ):
        mlflow.log_artifact(
            "artifacts/evaluation/precision_recall_curve.png"
        )

    if os.path.exists(
        "artifacts/evaluation/feature_importance.png"
    ):
        mlflow.log_artifact(
            "artifacts/evaluation/feature_importance.png"
        )

    if os.path.exists(
        "artifacts/evaluation/classification_report.txt"
    ):
        mlflow.log_artifact(
            "artifacts/evaluation/classification_report.txt"
        )

    # -------------------------
    # Save Model File
    # -------------------------
    mlflow.log_artifact(
        "models/best_model.pkl"
    )

print("MLflow experiment completed successfully.")
```
